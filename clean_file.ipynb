```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv(r"D:\project\smart_premium\data\train.csv", index_col="id")
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Identify numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
# Fill numerical columns with MEDIAN
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with MODE
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#this is use for streamlit purpose
output_path = r"D:\project\smart_premium\data\clean_file.csv"
df.to_csv(output_path)
```

Pipline method for clean data

```python
# -----------------------------------------------------------------------------------------#
#..........................................................................................#
# Preprocessing pipeline for clean data only, its not in use currently   ### NOT IN USE ###
# -----------------------------------------------------------------------------------------#
#..........................................................................................#

num_c = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))])                     


cat_c = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent"))])

preprocess_clean = ColumnTransformer([
    ("num", num_c, num_cols),              #Numerical features
    ("cat", cat_c, cat_cols)              #Categorical features 
])

# Fit the pipeline on the data and transform
processed_clean_data = preprocess_clean.fit_transform(df)

# We want it back as a DataFrame with column names
import numpy as np

# Get transformed column names
num_fea = num_cols
cat_fea = cat_cols
all_fea = np.concatenate([num_fea, cat_fea])

clean_df = pd.DataFrame(processed_clean_data, columns=all_fea)
clean_df.to_csv(r"D:\project\smart_premium\data\clean_data.csv", index=False)
#.......................................................................#
######################## CLEAN DATA SAVED TO CSV ########################
#.......................................................................#
```
